```python
class Solution:
    def maxMatrixSum(self, matrix):
        tsum=0
        cneg=0
        ms = float('inf')
        for row in matrix:
            for i in row:
                if i<0:
                    cneg+=1
                mii=abs(i)
                tsum+=mii
                ms=min(ms,mii)
        if cneg % 2 == 0:
            return tsum
        else:
            return tsum - 2 * ms
```
