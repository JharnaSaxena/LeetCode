```python
#optimized code
class Solution(object):
    def threeSum(self, nums):
        empty=[]
        nums.sort()

        for i in range(len(nums)):
            if i>0 and nums[i]== nums[i-1]:
                continue
            # 2 pointer approach
            start,end=i+1,len(nums)-1#cause it was out of index
            while start<end:
                sumo=nums[i]+nums[start]+nums[end]
                if sumo==0:
                    empty.append([nums[i],nums[start],nums[end]])
                    start+=1
                    end-=1
                    while start < end and nums[start]==nums[start-1]:
                        start+=1
                    while start < end and nums[end]==nums[end+1]:
                        end-=1

                elif sumo<0:
                    start+=1
                else:
                    end-=1
        return empty
'''
class Solution(object):
    def threeSum(self, nums):
        empty=[]
        nums.sort()
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                for k in range(j+1,len(nums)):
                    if nums[i] + nums[j] + nums[k] == 0:
                        tri=[nums[i],nums[j],nums[k]]
                        if tri not in empty:
                            empty.append(tri)
        return empty'''
```
