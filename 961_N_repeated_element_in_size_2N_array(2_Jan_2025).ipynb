```python
class Solution(object):
    def repeatedNTimes(self, nums):
        flag =0
        n=len(nums)*2
        if set(nums)==n+1:
            flag+=1
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                if nums[i]== nums[j]:
                    return nums[i]

```
