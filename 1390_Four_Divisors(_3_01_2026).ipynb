```python
class Solution(object):
    def sumFourDivisors(self, nums):
        number=0
        for j in nums:
            sumo=0
            count=0
            for i in range(1,int(j**0.5)+1):
                if j%i==0:
                    if j//i==i:
                        count+=1
                        sumo+=i
                    else:
                        count+=2
                        sumo+=i+(j//i)
            if count==4:
                number+=sumo
        return number
```
