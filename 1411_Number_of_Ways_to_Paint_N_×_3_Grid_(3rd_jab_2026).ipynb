```python

class Solution(object):
    def numOfWays(self, n):
            modulo = 10**9 + 7
            a=6 #abc,acb,bac,bca,cab,cba by the 3 factorial
            b=6 # by factorials
            for _ in range(2, n + 1):
                aa=(2*a+2*b)%modulo
                bb=(2*a+3*b)%modulo
                a,b=aa,bb
            return (a+b)%modulo
```
